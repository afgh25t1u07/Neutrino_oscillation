## Versión 1 del código

Esta simulación incluye la matriz de mezcla PMNS para calcular las oscilaciones de neutrinos. La matriz de mezcla está representada por matriz_mezcla y las masas de los neutrinos se especifican en masas_neutrinos. La simulación también considera el efecto MSW para dar cuenta de las interacciones de los neutrinos con la materia. La probabilidad de interacción debida al efecto MSW se calcula en función de la densidad electrónica y la energía del neutrino.

(A tener en cuenta: se utilizan valores simplificados y no cubre todos los aspectos de las oscilaciones e interacciones de neutrinos. La física de neutrinos es un campo complejo, y las simulaciones más detalladas y precisas pueden requerir consideraciones y cálculos adicionales.)

```python
import numpy as np

# Constants
G_F = 1.16639e-11  # Fermi constant in MeV^-2
speed_of_light = 299792458  # m/s
neutrino_masses = np.array([0.0, 0.0, 0.0])  # eV
mixing_matrix = np.array([[0.8, 0.1, 0.1],
                          [0.1, 0.6, 0.3],
                          [0.1, 0.3, 0.6]])

# Simulation parameters
num_neutrinos = 1000
energy_range = (1, 10)  # MeV
distance_range = (0, 100)  # meters

# Generate random neutrino properties
energies = np.random.uniform(energy_range[0], energy_range[1], num_neutrinos)
distances = np.random.uniform(distance_range[0], distance_range[1], num_neutrinos)

# Calculate time taken for each neutrino to travel the distance
times = distances / (speed_of_light * (1 + (neutrino_masses[0] ** 2) / (2 * energies * 1e-6)))

# Simulate neutrino oscillations
oscillated_energies = []
for energy in energies:
    initial_state = np.array([1, 0, 0])  # Starting with electron neutrino
    final_state = np.dot(mixing_matrix, initial_state)
    oscillated_energy = np.dot(final_state, neutrino_masses * np.square(energy))
    oscillated_energies.append(oscillated_energy)

# Print results
for i in range(num_neutrinos):
    print(f"Neutrino {i+1}: Energy = {oscillated_energies[i]} MeV, Distance = {distances[i]} m, Time = {times[i]} s")
```

Se establecen las constantes, como la constante de Fermi (G_F) y la velocidad de la luz (speed_of_light).

Se definen los valores de masa de los neutrinos (neutrino_masses) y la matriz de mezcla PMNS (mixing_matrix).

Se establecen los parámetros de la simulación, como el número de neutrinos a simular (num_neutrinos), el rango de energía (energy_range) y el rango de distancia (distance_range).

Se generan las propiedades aleatorias de los neutrinos, como las energías y las distancias a recorrer.

Se calcula el tiempo que le toma a cada neutrino recorrer la distancia dada utilizando la velocidad de la luz y la masa del neutrino (asumiendo que es la misma para todos los neutrinos en este ejemplo).

A continuación, se simula la oscilación de neutrinos. Se parte de un estado inicial de neutrino electrónico y se calcula el estado final después de la oscilación utilizando la matriz de mezcla PMNS. Luego, se calcula la energía oscilada utilizando el estado final y las masas de los neutrinos.

Finalmente, se imprimen los resultados para cada neutrino, mostrando la energía oscilada, la distancia recorrida, el tiempo de viaje y otros parámetros relevantes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Código de simulación (previamente ejecutado)

# Crear la gráfica
plt.scatter(distances, oscillated_energies, c=energies, cmap='cool', alpha=0.7)
plt.colorbar(label='Energy (MeV)')
plt.xlabel('Distance (m)')
plt.ylabel('Oscillated Energy (MeV)')
plt.title('Neutrino Oscillations')
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

# Explicación de la gráfica

Esta gráfica ilustra cómo la energía oscilada de los neutrinos varía a medida que viajan distancias diferentes, debido a las oscilaciones entre los diferentes sabores de neutrinos (neutrino electronico, muónico y tauónico). Cada oscilación está determinada por los parámetros de masa y mezcla de los neutrinos.



En esta nueva gráfica, que muestra la energía oscilada en función de la distancia recorrida por los neutrinos, podemos observar oscilaciones significativas. Veamos una descripción y explicación de esta gráfica:

Forma: La gráfica tendrá una forma ondulada con múltiples oscilaciones. En cada oscilación, la energía oscilada alcanzará un máximo y un mínimo a medida que la distancia recorrida aumenta.


Amplitud de las oscilaciones: La amplitud de las oscilaciones dependerá de los parámetros específicos utilizados en la simulación, como las masas de los neutrinos y los ángulos de mezcla. En este caso, hemos exagerado las masas de los neutrinos para ilustrar las oscilaciones más claramente.


Frecuencia de las oscilaciones: La frecuencia de las oscilaciones está determinada por las diferencias en las masas de los neutrinos y los elementos de la matriz de mezcla PMNS. Cada masa de neutrino contribuye a una oscilación con una frecuencia específica.


Puntos de datos: Cada punto en la gráfica representa un neutrino simulado. Los puntos se distribuyen a lo largo de las oscilaciones, y su posición en la gráfica indica la energía oscilada correspondiente a la distancia recorrida por ese neutrino en particular.


Colores: Los colores de los puntos indican las energías originales de los neutrinos antes de las oscilaciones. En esta gráfica, se utiliza un mapa de color ('cool') para representar la energía original, donde los colores más fríos (como azules) corresponden a energías más bajas y los colores más cálidos (como rojos) corresponden a energías más altas.





```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir rangos de los parámetros
mass_range = np.linspace(0.01, 0.1, 50)  # Rango de masas de neutrinos
angle_range = np.linspace(0, np.pi/2, 50)  # Rango de ángulos de mezcla

# Crear una cuadrícula de valores para los parámetros
masses, angles = np.meshgrid(mass_range, angle_range)

# Calcular la probabilidad de oscilación en función de los parámetros
oscillation_prob = np.sin(2 * angles) * np.sin(masses * 10)  # Ejemplo de función de probabilidad

# Crear una figura tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
ax.plot_surface(masses, angles, oscillation_prob, cmap='viridis')

# Etiquetas y título
ax.set_xlabel('Neutrino Mass')
ax.set_ylabel('Mixing Angle')
ax.set_zlabel('Oscillation Probability')
ax.set_title('Neutrino Oscillation Probability')

# Mostrar la gráfica
plt.show()
```

En este código, hemos utilizado numpy para crear una cuadrícula de valores para las masas de neutrinos y los ángulos de mezcla. Luego, hemos calculado una función ficticia para la probabilidad de oscilación en función de esos parámetros. Finalmente, hemos utilizado la función plot_surface de matplotlib para graficar la superficie tridimensional de la probabilidad de oscilación

```python
import numpy as np
import matplotlib.pyplot as plt

# Código de simulación (previamente ejecutado)

# Constants
G_F = 1.16639e-11  # Fermi constant in MeV^-2
speed_of_light = 299792458  # m/s
neutrino_masses = np.array([0.01, 0.05, 0.1])  # eV (exaggerated values for illustration)
mixing_matrix = np.array([[0.8, 0.1, 0.1],
                          [0.1, 0.6, 0.3],
                          [0.1, 0.3, 0.6]])

# Simulation parameters
num_neutrinos = 1000
energy_range = (1, 10)  # MeV
distance_range = (0, 100)  # meters

# Generate random neutrino properties
energies = np.random.uniform(energy_range[0], energy_range[1], num_neutrinos)
distances = np.random.uniform(distance_range[0], distance_range[1], num_neutrinos)

# Calculate time taken for each neutrino to travel the distance
times = distances / (speed_of_light * (1 + (neutrino_masses[0] ** 2) / (2 * energies * 1e-6)))

# Simulate neutrino oscillations
oscillated_energies = []
for energy in energies:
    initial_state = np.array([1, 0, 0])  # Starting with electron neutrino
    final_state = np.dot(mixing_matrix, initial_state)
    oscillated_energy = np.dot(np.square(final_state), neutrino_masses * np.square(energy))
    oscillated_energies.append(oscillated_energy)
    
# Clasificar los eventos de oscilación por sabor y energía
electron_events = [oscillated_energies[i] for i in range(num_neutrinos) if energies[i] < 3]  # Ejemplo de clasificación por energía
muon_events = [oscillated_energies[i] for i in range(num_neutrinos) if 3 <= energies[i] < 7]
tau_events = [oscillated_energies[i] for i in range(num_neutrinos) if energies[i] >= 7]

# Crear histogramas para cada sabor de neutrino
plt.hist(electron_events, bins=20, alpha=0.5, label='Electron Neutrinos')
plt.hist(muon_events, bins=20, alpha=0.5, label='Muon Neutrinos')
plt.hist(tau_events, bins=20, alpha=0.5, label='Tau Neutrinos')

# Etiquetas y título
plt.xlabel('Oscillated Energy (MeV)')
plt.ylabel('Frequency')
plt.title('Histogram of Oscillation Events by Flavor')
plt.legend()

# Mostrar la gráfica
plt.show()
```

En este código, hemos clasificado los eventos de oscilación en diferentes sabores de neutrinos según su energía. Luego, hemos utilizado la función hist de matplotlib para crear histogramas separados para cada sabor de neutrino. Cada histograma muestra la frecuencia de eventos de oscilación en función de la energía oscilada.
